# Pre-Processing

```python
#Now, we check for Duplicates In the DataSet
df.duplicated().sum()
```

```python
#We found 1 duplicate value and we are going to remove it now. 
df=df.drop_duplicates(keep="first")
```

```python
X = df.iloc[:,:-1] # We get all but the output column
y = pd.DataFrame(df["output"]) # We get output column

print("Shape of X:",X.shape,"\nShape of Y:",y.shape)
```

# Data Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.15 , random_state = 53)
```

```python
print("Shape of x train:",x_train.shape,"\nShape of y train:", y_train.shape)
print("Shape of x test:",x_test.shape,"\nShape of y test:", y_test.shape)
```

```python
# Importing metrics libraries of sklearn
from sklearn.metrics import accuracy_score , f1_score , recall_score, precision_score
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
accuracy_table = pd.DataFrame(columns=["Name", "Accuracy_score", "precision_score", "F1_score","recall_score"], dtype=object)
```

# Methodology 

## 1. Naive bayes

```python
from sklearn.naive_bayes import BernoulliNB

model1 = BernoulliNB().fit(x_train, y_train)
y_pred = model1.predict(x_test)

acc = round(accuracy_score(y_test, y_pred)*100, 2)
f1 = round(f1_score(y_test, y_pred)*100, 2)
prec = round(precision_score(y_test, y_pred)*100, 2)
recall = round(recall_score(y_test, y_pred)*100, 2)

print("Accuracy: ", acc, "%")
print("F1 accuracy: ", f1 , "%")
print("Precision: ", prec , "%")
print("Recall : ", recall , "%")

a_series = pd.Series(["Naive bayes", acc, f1, prec, recall], index = accuracy_table.columns)
accuracy_table = accuracy_table.append(a_series, ignore_index=True)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

## 2. Multi-layer Perceptron

```python
from sklearn.neural_network import MLPClassifier

model2 = MLPClassifier(activation='relu', learning_rate='adaptive', max_iter=500).fit(x_train, y_train)
y_pred = model2.predict(x_test)

acc = round(accuracy_score(y_test, y_pred)*100, 2)
f1 = round(f1_score(y_test, y_pred)*100, 2)
prec = round(precision_score(y_test, y_pred)*100, 2)
recall = round(recall_score(y_test, y_pred)*100, 2)

print("Accuracy: ", acc, "%")
print("F1 accuracy: ", f1 , "%")
print("Precision: ", prec , "%")
print("Recall : ", recall , "%")


a_series = pd.Series(["MLP", acc, f1, prec, recall], index = accuracy_table.columns)
accuracy_table = accuracy_table.append(a_series, ignore_index=True)
```

```python
confusion_matrix(y_test, y_pred)
```

## 3. Decision Tree Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

model3 = KNeighborsClassifier(n_neighbors=4).fit(x_train, y_train)
y_pred = model3.predict(x_test)

acc = round(accuracy_score(y_test, y_pred)*100, 2)
f1 = round(f1_score(y_test, y_pred)*100, 2)
prec = round(precision_score(y_test, y_pred)*100, 2)
recall = round(recall_score(y_test, y_pred)*100, 2)

print("Accuracy: ", acc, "%")
print("F1 accuracy: ", f1 , "%")
print("Precision: ", prec , "%")
print("Recall : ", recall , "%")


a_series = pd.Series(["DT", acc, f1, prec, recall], index = accuracy_table.columns)
accuracy_table = accuracy_table.append(a_series, ignore_index=True)
```

```python
confusion_matrix(y_test, y_pred)
```

## 4. Ensemble - Stacking 

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

estimators = [('nb', model1), ('mlp', model2), ('dt', model3)]

model4 = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression()).fit(x_train, y_train)
y_pred = model4.predict(x_test)

acc = round(accuracy_score(y_test, y_pred)*100, 2)
f1 = round(f1_score(y_test, y_pred)*100, 2)
prec = round(precision_score(y_test, y_pred)*100, 2)
recall = round(recall_score(y_test, y_pred)*100, 2)

print("Accuracy: ", acc, "%")
print("F1 accuracy: ", f1 , "%")
print("Precision: ", prec , "%")
print("Recall : ", recall , "%")

a_series = pd.Series(["Stacking", acc, f1, prec, recall], index = accuracy_table.columns)
accuracy_table = accuracy_table.append(a_series, ignore_index=True)
```

```python
confusion_matrix(y_test, y_pred)
```
